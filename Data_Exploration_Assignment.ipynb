2 Data manipulation in Pandas

```python
import pandas as pd
# Create a sales dataset using dictionary
sales = {"fruit": ["Apple","Banana","Pear"],
         "unit_price" : [2,1,1],
        "quantity": [10,20,5],}

# Convert dictionary to datafram
sales_df = pd.DataFrame(sales)
sales_df["price"] = sales_df["unit_price"]*sales_df["quantity"]
sales_df
```

This section create a sale dictionary and convert to dataframe.

3 Create a bar chart to visualized the dataframe.

```python
# Data visualization
import matplotlib.pyplot as plt
import seaborn as sns

#set a simple style
sns.set(style = "whitegrid")

#create a bar chart
plt.figure(figsize = (10,5))
sns.barplot(x = "fruit",y = "price",data = sales_df)
plt.title("Fruit sales",fontsize=14)
plt.xlabel("Fruit")
plt.ylabel("Price")
plt.bar(sales_df["fruit"],sales_df["price"])
plt.show()
```

A nice bar chart showing the salse of fruits.We can easyly identify the sales of different fruits.
